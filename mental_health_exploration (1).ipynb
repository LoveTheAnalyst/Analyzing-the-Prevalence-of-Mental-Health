#### **The Mission of the Project:**
To uncover trends about mental health and create awareness on the importance of taking care of our mental health.

#### **Abstract**
Mental health includes our emotional, psychological, and social well-being. It affects how we think, feel, and act. It also helps determine how we handle stress, relate to others, and make choices. Mental health is important at every stage of life, from childhood and adolescence through adulthood. (https://www.mentalhealth.gov/basics/what-is-mental-health). 
Several myths and facts have been attributed to mental health and it makes it difficult to identify and seek help early in situations of disorders. This project seeks to create awareness by showing causes and consequences of not attaching importance to healthy mental health.

The causes and consequences of mental health prpblems are classified by prevalence of the following:
1. Depression
2. Anxiety disorders
3. Bipolar disorder
4. Eating disorders (clinical anorexia & bulimia)
5. Schizophrenia
6. Drug use disorder (excluding alcohol)
7. Alcohol use disorder

Generally, collection and availability of mental health data is challenging especially in lower income countries due to lack of human resources, low penetration of technology, conflicts/war, financial constraints, lack of  modern data generation tools, amongst other reasons. These result in little or no awareness, less attention and treatment of mental disorders. Figures presented in this project should be taken as estimates of mental health disorder prevalence and not diagnosis.

#### Merge all the files horizontally

```python
#load the files to be joined 
alcohol_disorders_cleaned = pd.read_csv("alcohol_disorders_cleaned.csv")
anxiety_disorders_cleaned = pd.read_csv("anxiety_disorders_cleaned.csv")
bipolar_disorders_cleaned = pd.read_csv("bipolar_disorders_cleaned.csv")
depression_cleaned = pd.read_csv("depression_cleaned.csv")
drug_use_cleaned = pd.read_csv("drug_use_cleaned.csv")
eating_disorder_cleaned = pd.read_csv("eating_disorder_cleaned.csv")
schizophrenia_new_cleaned = pd.read_csv("schizophrenia_new_cleaned.csv")


df_all_cols = pd.concat([alcohol_disorders_cleaned, anxiety_disorders_cleaned, bipolar_disorders_cleaned, depression_cleaned,
                       drug_use_cleaned,eating_disorder_cleaned, schizophrenia_new_cleaned ], axis = 1)

# We use the axis=1 parameter to indicate that it is the columns that need to be joined together.
combined_disorders = df_all_cols
```

#### Limitation
Mental health disorders remain widely under-reported.

These result in little or no awareness, less attention and treatment of mental disorders.

#### Recommendation

Mental health issues in Nigeria can be reduced by:
- Creating awareness and educating the public about mental health
-  Patients having equal rights to treatment as individuals with physical illnesses.
- It can also be reduced by Integration of mental health into general health care services at all healthcare levels. 
- Promotion of healthy attitudes and positive socio-cultural attributes to aid positive mental health.
- Use of appropriate preventive therapeutic measures to reduce the problems of drug abuse
